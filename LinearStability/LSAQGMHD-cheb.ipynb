# Linear Stability: QG MHD Jet Model #

Here will compute the linear stability of the Bickley Jet for the one-layer quasi-geostrophic magnetohydrodynamic model.  


## Governing Equations

The nondimensional equations read
\begin{equation}
\begin{aligned}
    \partial_t q + J(\psi,q+\beta y) - M^2 J(A,\nabla^2A) - \frac{1}{R_e} \nabla^2 q= 0,\\
    \partial_t A + J(\psi,A)- \frac{1}{R_m} \nabla^2 A = 0,
\end{aligned}    
\end{equation}
with potential vorticity $q = \nabla^2 \psi - F \psi$ and where $F=(L/L_d)^2$ and $M = B/U$ are dimensionless parameters.

## Basic State
Assume that we have an exact steady solution that only depends on the meridional coordinate.  We define the streamfunction, velocity, Potential Vorticity and Magnetic Potential to be,

$$
\psi = \Psi(y), \quad\quad
u = U(y) = - \partial_y \Psi, \quad \quad
q = Q(y) =  \partial_{yy} \Psi - F\Psi \quad\quad
A = A(y) 
$$

We use the convention that captial letters are for the basic state. Notice that the components of the magnetic field are related to the magnetic potential by 

$$
B_x = -\partial_y A, \quad\quad B_y = \partial_x A
$$

such that if we consider a constant zonal magnetic field ($B_x = 1$), we require $A(y) = -y$.

## Perturbation
Next we assume that we have small amplitude perturbations that can depend on either spatial coordiante and time, $\psi'(x,y,t),$ $a'(x,y,t)$. If we substitute a perturbed basic state,

$$
\psi = \Psi(y)+\psi', \quad\quad A = A(y) + a',
$$

into the governing nonlinear equations, the smallness of the perturbations allows us to neglect the nonlinear terms.  Physically, that means we neglect the advection of the perturbation by the pertubation but still maintain the advection of the pertrubation by the mean state and the advection of the mean state by the perturbation.


## Linearized Equations

The linearized governing equation are then
\begin{equation}
\begin{aligned}
    \partial_t q' + J(\Psi, q'+\beta y) + J(\psi', Q+\beta y) - M^2J(A,\nabla^2a') - M^2J(a',\nabla^2A) - \frac{1}{R_e}\nabla^2q'= 0, \\
    \partial_t a' + J(\psi',A) + J(\Psi,a') - \frac{1}{R_m}\nabla^2a'= 0
\end{aligned}    
\end{equation}
which we can rewrite using the form of the basic state (indepedence of the zonal coordinate) and dropping primes for simplicity,
\begin{equation}
\begin{aligned}
    \partial_t q - (\partial_y\Psi)(\partial_xq) + (\partial_x\psi)(\partial_{y}Q +\beta) + M^2\left( (\partial_yA)(\partial_x\nabla^2a) - (\partial_xa)(\partial_y\nabla^2A)\right) - \frac{1}{R_e}\nabla^2q' = 0, \\
\partial_t a + \partial_x\psi\partial_yA - \partial_y\Psi\partial_xa - \frac{1}{R_m}\nabla^2a' = 0.
\end{aligned}    
\end{equation}
Using our previous definitions for the zonal velocity and magnetic field such that $U = -\partial_y \Psi$ and $B = -\partial_y A$, our equations finally read
\begin{equation}
\begin{aligned}
    \partial_t q +U \partial_x (\nabla^2 \psi) + (\beta-U_{yy})\partial_x\psi - M^2B\partial_x\nabla^2a + M^2B_{yy}\partial_xa - \frac{1}{R_e}\nabla^2q = 0,\\
    \partial_t a - B\partial_x\psi + U\partial_xa - \frac{1}{R_m}\nabla^2a= 0,
\end{aligned}    
\end{equation}
where the terms proportional to $F$ from the second and third terms cancel via: 
$$ 
-\partial_y\Psi(\partial_x\nabla^2\psi-F^2\partial_x\psi) + \partial_x\psi(\partial_y\nabla^2\Psi-F^2\partial_y\Psi) = U \partial_x (\nabla^2 \psi ) -U_{yy}\partial_x\psi.
$$

## Modal Decompositon
Since the domain is assumed to be periodic in the zonal direction and we are looking for wave-like solutions, we assume that the perturbation can be decomposed as,
\begin{equation}
\begin{aligned}
    \psi = \hat \psi(y) \exp[i (k x - \omega t)], \quad\quad a = \hat a(y) \exp[i (k x - \omega t)],
\end{aligned}    
\end{equation}
which in turn transforms the Laplacian operator such that $\hat\nabla^2 = \partial_{yy}-k^2$.
Substituting this into our linear equation and dividing through by the exponential, we find 
(multiplying by $i/k$ and using $\hat q = (\hat\nabla^2-F)\hat\psi$)
\begin{equation}
\begin{aligned}
    \left(\frac{\omega}{k}(\hat\nabla^2 - F) - U\hat\nabla^2 + (U_{yy}-\beta) -\frac{i}{k R_e}\hat\nabla^2(\hat\nabla^2-F) \right)\hat\psi +M^2\left( B\hat\nabla^2 -  B_{yy}\right)\hat a &= 0, \\
    \frac{\omega}{k}\hat a +B\hat\psi -(U + \frac{i}{k R_m}\hat\nabla^2)\hat a &= 0,
\end{aligned}    
\end{equation}
or in matrix form, (as a generalized EVP: $A{\bf v} = cB{\bf v}$, with $c=\omega/k$)
\begin{equation}
\begin{aligned}
    \begin{bmatrix}
        (U+\frac{i}{k R_e}(\hat\nabla^2 -F))\hat\nabla^2 -U_{yy}+\beta & -M^2(B\hat\nabla^2 -B_{yy}) \\
        -B & U+\frac{i}{k R_m}\hat\nabla^2 
    \end{bmatrix}
    \begin{bmatrix}
        \psi  \\
        a  
    \end{bmatrix}
     = c
    \begin{bmatrix}
        \hat\nabla^2 -F & 0 \\
        0 & 1 
    \end{bmatrix}
    \begin{bmatrix}
        \psi  \\
        a  
    \end{bmatrix}.
\end{aligned}    
\end{equation}
In the absence of any diffusion, the system reduces to 
\begin{equation}
\begin{aligned}
    \begin{bmatrix}
        U\hat\nabla^2 -U_{yy}+\beta & -M^2(B\hat\nabla^2 -B_{yy}) \\
        -B & U 
    \end{bmatrix}
    \begin{bmatrix}
        \psi  \\
        a  
    \end{bmatrix}
     = c
    \begin{bmatrix}
        \hat\nabla^2 -F & 0 \\
        0 & 1 
    \end{bmatrix}
    \begin{bmatrix}
        \psi  \\
        a  
    \end{bmatrix}.
\end{aligned}    
\end{equation}
Note: $Uk^2+U_{yy} = Uk^2+UF - (UF +(-U_y)_y) = U(k^2+F) - Q_y$.

## Numerical Approach
This is a generalized differential eigenvalue problem where the eigenvalue is the phase speed, $\omega/k$, and the eigenvector determines the spatial structure.  For a general velocity field it is impossible to find an analytical solution to this problem.  However, if we discretize the meridional domain and then approximate the differential operators with algebraic operators, we then obtain an algebraic eigenvalue problem that is much easier to solve. 

There are a variety of different ways to approximate the derivatives.  The most straight forward is to use a second order centre differencing scheme.  This is adequate for many problems but only second order accurate.  A method that is usually far better is to use a spectral method.  Since our perturbation need not be periodic in the meridional direction, it is perhaps best to use a Chebyschev grid in $y$.  Reference Trefethen.

## Differential Operators
If we define $L$ to be the length in the $y$ direction and $N$ is the number of points in that direction, then a second-order Finite Difference operator can be built as follows:

```python
import numpy as np
import scipy.linalg as spalg
import matplotlib
import matplotlib.pyplot as plt
import sys

try: # Try using mpi
    from mpi4py import MPI
    comm = MPI.COMM_WORLD
    rank = comm.Get_rank()
    num_procs = comm.Get_size()
except:
    rank = 0
    num_procs = 1
print("Proc {0:d} of {1:d}".format(rank+1,num_procs))

# Physical parameters
L     = 8*np.pi                      # Length of domain
N     = 512                          # Number of grid points
F     = 0.0                          # Froude number
M2    = 1e-2                         # Magnetic number
beta  = 0                            # nondim beta from coriolis
Re    = 1e4                          # Reynolds Number
Pm    = 1.0                          # Prandtl Number
Rm    = Pm*Re                        # Magnetic Reynolds Number

Reinv = 1/Re                         # Inverse Reynolds number to allow for inviscid case
Rminv = 1/Rm                         # Inverse Reynolds number to allow for inviscid case

# Jet parameters
Lj = 1.0                             # width of jet
Uj = 1.0                             # maximum velocity of jet

kcrit = 2*np.pi/L                    # critical wavenumber (multiples fit in the domain)

method = 'cheb'

if rank==0:
    print("Computational Parameters:")
    print("Domain Length (L) = "+str(L))
    print("Domain Num Points = "+str(N))
    print("Domain Resolution = "+str(L/N))
    print("")
    print("Physical Parameters")
    print("F  = "+str(F))
    print("M2 = "+str(M2))
    print("Re = "+str(Re))
    print("Rm = "+str(Rm))
```

```python
## FD2 computes the 2nd-order Finite Difference Operator
## ------
#    matrix on N+1 points (i.e. N intervals)
#    D2 = differentiation matrix
#    y  = uniform grid

def FD2(N,L):

    y  = np.linspace(0, L, N+1)       # Meridional coordiante
    dy = y[1] - y[0]                  # grid spacing
    dy2 = dy**2                       # square of dy

    Dyy = np.diag(1/dy2*np.ones(N),-1)   \
        + np.diag(-2/dy2*np.ones(N+1),0) \
        + np.diag(1/dy2*np.ones(N),1) 
    Dyy[0,0:3] = np.array([1, -2, 1]/dy2)
    Dyy[N,N-2:N+1] = np.array([1, -2, 1]/dy2)
    
    return y,Dyy
```

```python
## CHEB computes the Chebyshev differentiation matrix
## ------
#    matrix on N+1 points (i.e. N intervals)
#    D = differentiation matrix
#    x = Chebyshev grid

def cheb(N):
    if N == 0:
        D = 0
        x = 1
    else:
        x = np.cos(np.pi*np.array(range(0,N+1))/N).reshape([N+1,1])
        c = np.ravel(np.vstack([2, np.ones([N-1,1]), 2])) \
            *(-1)**np.ravel(np.array(range(0,N+1)))
        c = c.reshape(c.shape[0],1)
        X = np.tile(x,(1,N+1))
        dX = X-(X.conj().transpose())
        D  = (c*(1/c).conj().transpose())/(dX+(np.eye(N+1)))   # off-diagonal entries
        D  = D - np.diag(np.sum(D,1))   # diagonal entries
    return D,x
## ------
```

```python
if method == 'FD2':

    ## FD2
    y,Dyy = FD2(N,L)

elif method == 'cheb':
    
    ## Cheb
    Dy,y = cheb(N)

    y   = (y[:,0]+1)*L/2
    Dy  = Dy*(2/L)
    Dyy = np.dot(Dy,Dy)

else:
    
    print("Method must be either FD2 or cheb")
    sys.exit()
```

```python
# Define Useful Operators
I  = np.identity(N-1)
O  = np.zeros([N-1,N-1])

# Define Basic State
P  = -Uj*np.tanh((y-L/2)/Lj)
U  = Uj/(np.cosh((y-L/2)/Lj))**2
#P  = np.cos(y)
#U  = np.sin(y)
Q  = -F*P + np.dot(Dyy,P)  #+ beta*y
#Q2 = -2.*Uj*np.tanh((y-L/2)/Lj)/(np.cosh((y-L/2)/Lj))**2
Uyy= np.dot(Dyy,U)
B0 = np.ones(np.shape(y)) 
Byy= np.dot(Dyy,B0)
```

```python
# Plot Basic State
plt.figure(figsize=(12,8))
plt.clf()

plt.subplot(411)
plt.plot(y, P,'-b')
plt.grid(True)
plt.xlim([0, L])
plt.xlabel('y')
plt.title('Zonal Streamfunction')

plt.subplot(412)
plt.plot(y, U,'-b')
plt.grid(True)
plt.xlim([0, L])
plt.xlabel('y')
plt.title('Zonal Velocity')

plt.subplot(413)
plt.plot(y, Q,'-b')
plt.grid(True)
plt.xlim([0, L])
plt.xlabel('y')
plt.title('Zonal PV')

plt.subplot(414)
plt.plot(y, B0,'-b')
plt.grid(True)
plt.xlim([0, L])
plt.xlabel('y')
plt.title('Zonal Mag Field')

plt.tight_layout()
plt.show()
```

```python
# Define range of wavenumbers 
dk = 2e-2
#kk = np.arange(dk,1+dk,dk)
kk = np.array([3/4])
Nk = len(kk)

# Define storage vectors: DIM = [psi, a]x[num modes]x[num wavenumbers] = 2 x Ne x Nk
Ne = 10
c_vals = np.zeros((Ne,Nk),dtype=complex)
grow = np.zeros((Ne,Nk))
freq = np.zeros((Ne,Nk))
modes = np.zeros((2,N+1,Ne,Nk),dtype=complex)

# Loop over wavenumbers
cnt = 0
for cnt in range(rank,len(kk),num_procs):

    k = kk[cnt]
    k2 = k**2
    nabla = Dyy[1:-1,1:-1] - k2*I
    
    # DIFFUSION
    B = np.vstack((np.hstack((nabla - F*I, O)),
                  np.hstack((O, I))))
    A = np.vstack((np.hstack((np.dot((np.diag(U[1:-1],0)+1j*Reinv/k*(nabla - F*I)),nabla) - np.diag(Uyy[1:-1],0) +beta*I, 
                             -M2*(np.dot(np.diag(B0[1:-1],0),nabla) - np.diag(Byy[1:-1],0)))),
                  np.hstack((-np.diag(B0[1:-1],0), 
                             np.diag(U[1:-1],0)+1j*Rminv/k*(nabla)))))    
    
    # NO DIFFUSION, CLASSICAL
    #B = np.vstack((np.hstack((nabla - F*I, O)),
    #              np.hstack((O, I))))
    #A = np.vstack((np.hstack((np.dot(np.diag(U[1:-1],0),nabla) - np.diag(Uyy[1:-1],0) +beta*I,0*I)),
    #              np.hstack((-np.diag(B0[1:-1],0),np.diag(U[1:-1],0))))) 

    # Solve for eigenvalues
    eigVals,eigVecs = spalg.eig(A,B)

    # Sort eigenvalues and eigenvectors
    ind = (-np.imag(eigVals)).argsort()
    
    eigVecs = eigVecs[:,ind]
    eigVals = k*eigVals[ind]

    # Store eigenvalues and eigenvectors
    c_vals[:,cnt] = eigVals[0:Ne]/k
    grow[:,cnt] = eigVals[0:Ne].imag
    freq[:,cnt] = eigVals[0:Ne].real
    modes[0,1:N,:,cnt] = eigVecs[0:N-1,0:Ne]
    modes[1,1:N,:,cnt] = eigVecs[N-1:2*N,0:Ne]
    
    print ('Wavenumber (', int(cnt+1), '/', int(Nk),')', ': ',"{:.2f}".format(k),', Growth: ',"{:.4f}".format(grow[0,cnt]),', Phase: ',"{:.4f}".format(freq[0,cnt]))
    cnt += 1

c_vals[:,:] = comm.reduce(c_vals[:,:],op=MPI.SUM, root=0)    
modes[:,:,:] = comm.reduce(modes[:,:,:],op=MPI.SUM, root=0)
grow[:,:] = comm.reduce(grow[:,:],op=MPI.SUM, root=0)    
freq[:,:] = comm.reduce(freq[:,:],op=MPI.SUM, root=0)

print("Done!")
```

```python
if rank==0:
    # Find max growth rates
    Imax = np.zeros((Ne,1), dtype=int)
    for ii in range(Ne):
        Imax[ii] = np.argmax(grow[ii,:])
        print("Max growth for curve", ii+1, "is", grow[ii,Imax[ii]])

    # Plot the two most unstable modes
    plt.figure(figsize=(10,10))
    plt.clf()
    plt.plot(kk,grow[0,:],'.b')
    plt.plot(kk[Imax[0]],grow[0,Imax[0]],'ob', markersize=12)
    plt.plot(kk,grow[1,:],'.r')
    plt.plot(kk[Imax[1]],grow[1,Imax[1]],'or', markersize=12)
    plt.plot(kk,grow[2,:],'.g')
    plt.plot(kk[Imax[2]],grow[2,Imax[2]],'og', markersize=12)
    plt.plot(kk,grow[3,:],'.y')
    plt.plot(kk[Imax[3]],grow[3,Imax[3]],'oy', markersize=12)
    plt.grid(True)
    plt.ylim([0,1.1*np.amax(grow[0,:])])
    plt.title("Growth Rates of QGMHD (spectral) Jet: "+"M2 = "+str(M2)+", F = "+str(F))
    #plt.legend(loc='best')
    plt.tight_layout()
    plt.show()
    #plt.savefig("QGMHD_growth_bickley_spectral.png") 
    
    
```
